# Bayesian Logistic Regression with ESS and MESS

This notebook implements Bayesian logistic regression using Elliptical Slice Sampling (ESS) and MESS (Multiple-proposal ESS) from the original Murray et al. (2010) paper. We compare ESS, MESS uniform, and MESS with different distance metrics (angular, euclidean) on a synthetic binary classification problem.

Key metrics tracked:
- Number of shrinking steps (intervals)
- Effective Sample Size (ESS)
- Mean Squared Jumping Distance (MSJD)
- Computation time

## Import Required Libraries

```python
import sys
import os
import time

# Get absolute path to src directory (go up from notebooks to repo root)
repo_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
src_path = os.path.join(repo_root, 'src')
sys.path.insert(0, src_path)

print(f"Repo root: {repo_root}")
print(f"Added to path: {src_path}")

import numpy as np
import matplotlib.pyplot as plt
from multiproposal.algorithms.ess import ess_step
from multiproposal.algorithms.mess import mess_step
from multiproposal.algorithms.effective_sample_size import (
    estimate_effective_sample_size,
    compute_mean_squared_jumping_distance,
    compute_normalized_jumping_distance
)
from multiproposal.problems.logistic_regression import BayesianLogisticRegression
from multiproposal.data.logistic_regression import generate_logistic_regression_data
```

## Generate Synthetic Logistic Regression Data

```python
# Generate synthetic logistic regression data
data = generate_logistic_regression_data(
    n_samples=100,
    n_features=10,
    seed=42
)

X = data['X']
y = data['y']
beta_true = data['beta_true']

# Create problem
problem = BayesianLogisticRegression(X, y, prior_var=1.0)

print(f"Data shapes:")
print(f"  X: {X.shape}")
print(f"  y: {y.shape}")
print(f"  Number of class 0: {np.sum(y == 0)}")
print(f"  Number of class 1: {np.sum(y == 1)}")
print(f"\nTrue coefficients: {beta_true}")

# Initialize sampling
beta0 = problem.sample_prior(np.random.default_rng(seed=42))
print(f"\nInitial log-likelihood: {problem.log_likelihood(beta0):.4f}")
```

## Experiment Setup

```python
# Sampler parameters
n_iters = 10000
burn_in = 1000
seed = 42

# MESS M values to test
M_values = [1, 2, 5, 10, 15, 20, 25, 30, 40, 50, 75, 100, 200]
M_values_filtered = [M for M in M_values if M <= 100]

print(f"Experiment setup:")
print(f"  n_iters: {n_iters}")
print(f"  burn_in: {burn_in}")
print(f"  M values: {M_values}")
```

## Run MCMC Sampling Experiments

Run MESS with uniform, angular, and euclidean variants for different M values.

```python
# Store results for all variants
results = {
    'uniform': {'intervals': {}, 'chains': {}, 'times': [], 'statistics': {}},
    'angular': {'intervals': {}, 'chains': {}, 'times': [], 'statistics': {}},
    'euclidean': {'intervals': {}, 'chains': {}, 'times': [], 'statistics': {}},
    'ess': {'uniform': {}, 'angular': {}, 'euclidean': {}},
    'msjd': {'uniform': {}, 'angular': {}, 'euclidean': {}},
    'log_likelihood': {'uniform': {}, 'angular': {}, 'euclidean': {}},
}

# Run MESS Uniform
print(f"Running MESS Uniform...")
for M in M_values:
    print(f"  M = {M}")
    rng = np.random.default_rng(seed)
    chain = np.zeros((n_iters, X.shape[1]))
    intervals = np.zeros(n_iters, dtype=int)
    x = beta0.copy()

    t0 = time.time()
    for t in range(n_iters):
        x, nr_intervals, _ = mess_step(x, problem, rng, M=M, use_lp=False)
        chain[t] = x
        intervals[t] = nr_intervals
    elapsed = time.time() - t0

    results['uniform']['chains'][M] = chain
    results['uniform']['intervals'][M] = intervals
    results['uniform']['times'].append(elapsed)

    # Compute statistics
    intervals_post = intervals[burn_in:]
    results['uniform']['statistics'][M] = {
        'mean_intervals': np.mean(intervals_post),
        'std_intervals': np.std(intervals_post),
        'median_intervals': np.median(intervals_post),
    }

    print(f"    Time: {elapsed:.2f}s, Mean intervals: {np.mean(intervals_post):.4f}")

# NOTE: Angular/Euclidean runs are disabled for uniform-only analysis.
#
# # Run MESS Angular
# print(f"\nRunning MESS Angular...")
# for M in M_values_filtered:
#     print(f"  M = {M}")
#     rng = np.random.default_rng(seed)
#     chain = np.zeros((n_iters + 1, X.shape[1]))
#     chain[0] = beta0.copy()
#     intervals = np.zeros(n_iters, dtype=int)
#     x = beta0.copy()
#
#     t0 = time.time()
#     for t in range(n_iters):
#         x, nr_intervals, _ = mess_step(x, problem, rng, M=M, use_lp=True,
#                                        distance_metric='angular', lam=0.05)
#         chain[t + 1] = x
#         intervals[t] = nr_intervals
#     elapsed = time.time() - t0
#
#     results['angular']['chains'][M] = chain
#     results['angular']['intervals'][M] = intervals
#     results['angular']['times'].append(elapsed)
#
#     # Compute statistics
#     intervals_post = intervals[burn_in + 1:]
#     results['angular']['statistics'][M] = {
#         'mean_intervals': np.mean(intervals_post),
#         'std_intervals': np.std(intervals_post),
#         'median_intervals': np.median(intervals_post),
#     }
#
#     print(f"    Time: {elapsed:.2f}s, Mean intervals: {np.mean(intervals_post):.4f}")
#
# # Run MESS Euclidean
# print(f"\nRunning MESS Euclidean...")
# for M in M_values_filtered:
#     print(f"  M = {M}")
#     rng = np.random.default_rng(seed)
#     chain = np.zeros((n_iters + 1, X.shape[1]))
#     chain[0] = beta0.copy()
#     intervals = np.zeros(n_iters, dtype=int)
#     x = beta0.copy()
#
#     t0 = time.time()
#     for t in range(n_iters):
#         x, nr_intervals, _ = mess_step(x, problem, rng, M=M, use_lp=True,
#                                        distance_metric='euclidean', lam=0.05)
#         chain[t + 1] = x
#         intervals[t] = nr_intervals
#     elapsed = time.time() - t0
#
#     results['euclidean']['chains'][M] = chain
#     results['euclidean']['intervals'][M] = intervals
#     results['euclidean']['times'].append(elapsed)
#
#     # Compute statistics
#     intervals_post = intervals[burn_in + 1:]
#     results['euclidean']['statistics'][M] = {
#         'mean_intervals': np.mean(intervals_post),
#         'std_intervals': np.std(intervals_post),
#         'median_intervals': np.median(intervals_post),
#     }
#
#     print(f"    Time: {elapsed:.2f}s, Mean intervals: {np.mean(intervals_post):.4f}")

print(f"\nAll experiments completed!")
```

## Figure 2c: Computational Cost vs M

Plot the computational cost (shrinking steps × likelihood evaluation cost × M) as a function of M, showing serial and parallelized (theoretical) costs.

```python
# Estimate cost per likelihood evaluation from M=1 case
# For M=1, total_time = n_iters * (mean_intervals_M1 + 1) * cost_per_lik_eval
M_idx_1 = M_values.index(1)
intervals_post_M1 = results['uniform']['intervals'][1][burn_in:]
mean_intervals_M1 = float(np.mean(intervals_post_M1))
time_M1 = results['uniform']['times'][M_idx_1]
cost_per_lik_eval = time_M1 / (n_iters * (mean_intervals_M1 + 1.0))

print(f"Estimated cost per likelihood evaluation: {cost_per_lik_eval:.6f} seconds")

# Compute computational costs
M_plot_cost = [M for M in M_values if M <= 50]
serial_cost = []
parallel_cost = []

for M in M_plot_cost:
    intervals_post = results['uniform']['intervals'][M][burn_in:]
    mean_intervals = float(np.mean(intervals_post))
    
    # Serial: mean_intervals × cost_per_lik_eval × M
    serial = mean_intervals * cost_per_lik_eval * M
    # Parallelized: mean_intervals × cost_per_lik_eval
    parallel = mean_intervals * cost_per_lik_eval
    
    serial_cost.append(serial)
    parallel_cost.append(parallel)
    print(f"  M = {M}: serial cost = {serial:.6f}s, parallel cost = {parallel:.6f}s")

# Plot computational costs
fig, ax = plt.subplots(figsize=(10, 7))

# Grayscale and different line styles/widths
ax.plot(
    M_plot_cost,
    serial_cost,
    marker='s',
    markersize=7,
    linewidth=2.0,
    linestyle='-.',
    color='gray',
    label='Serial computational cost'
)
ax.plot(
    M_plot_cost,
    parallel_cost,
    marker='^',
    markersize=7,
    linewidth=1.5,
    linestyle='--',
    color='black',
    label='Parallelized computational cost (theoretical)'
)

ax.set_xlabel('M (Number of Proposals)', fontsize=16)
ax.set_ylabel('Computational cost (seconds)', fontsize=16)
ax.set_title('Computational Cost vs M', fontsize=18)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=12, loc='best')
ax.tick_params(labelsize=12)

plt.tight_layout()
fig.savefig('log_regression_computational_cost.png', dpi=600)
plt.show()
```
